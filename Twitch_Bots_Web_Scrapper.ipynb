```python
#Importing the necessary libraries

import selenium
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import requests
import bs4
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
#Launch Chrome driver 
path = 'chromedriver.exe'
driver = webdriver.Chrome(path)
```

```python
# Enter to the site
driver.get("https://twitchinsights.net/bots")
time.sleep(2)
```

```python
#Scroll down
driver.execute_script("window.scrollTo(0, 1400)")
```

```python
#Select all entries to get the total bots identified
all_entries = WebDriverWait(driver, timeout=15).until(EC.element_to_be_clickable((By.XPATH,"/html/body/div[3]/div/div[2]/div/div/div[1]/div[2]/div/label/select/option[8]")))
all_entries.click()
```

```python
#Get the table containing the information of the bots identified
bots_table = driver.find_element(By.XPATH, "/html/body/div[3]/div/div[2]/div/div/div[2]/div")
```

```python
#Get the total number of bots identified
total_bots = int(bots_table.find_element(By.XPATH, "/html/body/div[3]/div/h2[2]/small").text.replace("(", "").replace(")", ""))

print(total_bots)
```

```python
#Define the information to be collected
usernames = []
channels = []
last_seen = []
links = []


#Create a loop to scrap the data of the total bots identified in the webpage
for i in range(1,total_bots+1):
    xpath = f"/html/body/div[3]/div/div[2]/div/div/div[2]/div/table/tbody/tr[{i}]"
    
    usernames.append(bots_table.find_element(By. XPATH,xpath).text.split(" ")[0])
    channels.append(bots_table.find_element(By. XPATH,xpath).text.split(" ")[1])
    last_seen.append(bots_table.find_element(By. XPATH,xpath).text.split(" ")[2:])
    
    #Get the links of the user
    
    users_info = bots_table.find_element(By. XPATH, xpath)
    profiles = users_info.find_element("tag name","a")
    links.append(profiles.get_attribute("href"))
    
    print(f"Scrapping bot number [{i}]")
```

```python
#Create a df
df = pd.DataFrame({"usernames": usernames, 
                  "links": links,
                  "channels": channels, 
                  "last_seen_list": last_seen})
```

### Cleaning the dataset

```python
# Define function to join elements of last_seen_list
def join_list_elements(lst):
    return ' '.join(lst)

# Apply function to each element of the last_seen_list column
df['last_seen_str'] = df['last_seen_list'].apply(join_list_elements)

#Drop last_seen_list column
df = df.drop('last_seen_list', axis=1)
```

```python
# Define the date format
date_format = "%a, %d %b %Y %H:%M:%S %Z"

# Define a function to convert a single date string to a datetime object
def convert_to_datetime(date_str):
    return datetime.strptime(date_str, date_format)

# Convert the column to datetime and assign back to the original column
df['last_seen'] = df['last_seen_str'].apply(convert_to_datetime)
```

```python
# Remove commas from the strings and convert to integers using astype()
df['channels'] = df['channels'].str.replace(',', '').astype(int)
```

```python
# Create a csv file with the data saved
df.to_csv('Twitch_bots.csv', index=False)
```
